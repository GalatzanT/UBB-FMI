```python
from random import randint

def are_duplicate_birthdays(n):
    zile_nastere = [randint(1, 365) for _ in range(n)]
    return len(zile_nastere) != len(set(zile_nastere))

def probabilitate_zi_de_nastere(n, nr_incercari=1000):
    cazuri_favorabile = 0

    for _ in range(nr_incercari):
        if are_duplicate_birthdays(n):
            cazuri_favorabile += 1


    probabilitate = cazuri_favorabile / nr_incercari
    return probabilitate


n = int(input("Introduceți numărul de persoane: "))
probabilitate = probabilitate_zi_de_nastere(n)
print(f"Probabilitatea ca într-un grup de {n} persoane să existe cel puțin două persoane cu aceeași zi de naștere este {probabilitate:.4f}")

```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist

n=5000
i=0
count=0;
axis('square')
axis((0, 1, 0, 1))
while(i<n):

  F=[random(),random()]
  if dist(F,(0.5,0.5))<0.5:
      plot(F[0],F[1],'ro')
      count+=1
  else:
      plot(F[0],F[1],'bo')
  i+=1
show()
```
