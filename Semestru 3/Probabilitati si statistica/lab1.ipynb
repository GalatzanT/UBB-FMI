```python
def aranjamente(cuv, nr, numar_total=False, aleator=False):
  aranjs = [a for a in permutations(cuv, nr)]
  if aleator==True:
    return sample(cuv,nr)
  elif numar_total==True:
    return len(aranjs)
  else:
    return list(aranjs)


print(aranjamente('word',2))
```
